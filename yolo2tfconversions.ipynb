## Convert Tensorflow (frozen inference graph) .pb  to tflite

```python
import tensorflow as tf
gf = tf.GraphDef()   
m_file = open('weights/yolov3.pb','rb')
gf.ParseFromString(m_file.read())

with open('somefile.txt', 'a') as the_file:
    for n in gf.node:
        the_file.write(n.name+'\n')

file = open('somefile.txt','r')
data = file.readlines()
print("output name = ")
print(data[len(data)-1])

print("Input name = ")
file.seek ( 0 )
print(file.readline())
```

    The above code creates somefile.txt file. 
    
    open it

    Take first line  and last but one line as input and output array names to   
    pass into tflite converter below.

```python
graph_def_file = "weights/yolov3.pb"
input_arrays = ["Const"]               # 
output_arrays = ["Concat_252/axis"]

converter = tf.lite.TFLiteConverter.from_frozen_graph(
        graph_def_file, input_arrays, output_arrays)
tflite_model = converter.convert()
open("converted_model.tflite", "wb").write(tflite_model)
```
